```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drug_times_count = pd.read_csv("drug_times_count.csv")
drugs_people_count = pd.read_csv("drug_people_count.csv")
people_drug_variety = pd.read_csv("people_drug_variety.csv")
```

```python
drug_times_count.head()
```

```python
plt.figure(figsize=(3, 9))
boxplot = drug_times_count.boxplot(column="times_taken", whis=[0, 100], widths=0.3)

# Remove top, right, and bottom spines
for spine in ["top", "right", "bottom"]:
    boxplot.spines[spine].set_visible(False)

# Remove x-axis ticks and labels
plt.xticks([])
plt.yticks(fontsize=12)
plt.xlabel("")
plt.ylabel("How Many Times Each Drug is Taken", fontsize=15)
plt.yscale("log")  # Using log scale for better visualization
plt.show()
```
